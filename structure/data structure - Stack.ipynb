## 스택 자료구조

1. 대표적인 후입선출 자료구조로 먼저 입력된 자료가 가장 늦게 반환
2. 스택의 자료구조로는 push, pop, top, size, empty 등의 기능을 제공
    - push : 스택에 값을 저장
    - pop : 가장 최근에 저장한 값을 제거
    - top : 스택의 최상단에 위치한 값을 확인
    - size : 스택에 몇 개의 원소가 입력되어있는지 확인
    - empty : 스택에 값이 있는지 없는지를 확인
3. Class 구조가 아닌 단순 list 연산으로 스택을 구현할 수도 있음

```python
class Stack:
    def __init__(self):
        self.data = []
        
    def push(self, input):
        self.data.append(input)
        
    def pop(self):
        if not self.data:
            return -1
        
        return self.data.pop()
        
    def top(self):
        if not self.data:
            return -1
        
        return self.data[-1]
    
    def size(self):
        
        return len(self.data)
    
    def empty(self):
        if not self.data:
            return 1
        
        return 0
```

```python
my_stack = Stack()
```

```python
my_stack.push('test')
```

```python
my_stack.top()
```

```python
my_stack.push('test2')
```

```python
my_stack.size()
```
